### Get rid of invalid zip codes

```python
zip_threshed = demographics.loc[demographics["zipcode"] >= 10000]
display(zip_threshed)
age_range = zip_threshed.iloc[:,2:15]
# display(age_range)
```

### Age groups => percentages/distribution

```python
# converts age ranges to percentages
total_pop = zip_threshed.iloc[:,1]
age_percentages = age_range.div(total_pop, axis='rows')
display(age_percentages)
sanity = age_percentages.sum(axis=1, skipna=None)
display(sanity) #sanity check to determine if they add to 1
```

### Include mean and median household income

```python
income = zip_threshed.iloc[:,24:26]
display(income)
```

### Make it neat

```python
zipcode = zip_threshed.iloc[:,0]
display(zipcode)
```

```python
cleaned_demographic = pd.concat([pd.DataFrame(zipcode), pd.DataFrame(age_percentages), pd.DataFrame(income)],axis = 1)
display(cleaned_demographic)
# cleaned_demographic.set_index('zipcode', inplace=True)
# display(cleaned_demographic)
```

### There are two metrics. The non-ZHVI have no useful data. Be gone

```python
re_ZHVI = real_estate.loc[real_estate["type"] == "ZHVI"]
re_recent_zlw = re_ZHVI.iloc[:,249:265]
display(re_recent_zlw)
```

### Get mean and percent change in Zillow rating of zipcode

```python
re_mean = re_recent_zlw.mean(axis = 1)
display(re_mean)
re_slope = (re_recent_zlw.iloc[:,11]-re_recent_zlw.iloc[:,0])/re_recent_zlw.iloc[:,11]
display(re_slope)
```

### Make it neat

```python
re_data = pd.concat([pd.DataFrame(re_ZHVI["zipcode"]),pd.DataFrame(re_ZHVI["size_rank"]),pd.DataFrame(re_ZHVI["state"]),pd.DataFrame(re_mean,columns = ["2016-2017_mean_zillow_score"]),pd.DataFrame(re_slope,columns = ["2016-2017_%change_zillow_score"])],axis = 1)
display(re_data)
```

```python
# match zipcodes in real estate dataset and demographic dataset.
cleaned_demographic.set_index('zipcode', inplace=True)
re_data.set_index("zipcode", inplace=True)
```

```python
display(cleaned_demographic)
display(re_data)
```

## Save Zipcodes that appear in both datasets

```python
idx1 = pd.Index(list(cleaned_demographic.index))
idx2 = pd.Index(list(re_data.index))
idx_int = idx1.intersection(idx2)
```

```python
re_data_final = re_data[re_data.index.isin(idx_int)]
dem_data_final = cleaned_demographic[cleaned_demographic.index.isin(idx_int)]
display(re_data_final.sort_index())
display(dem_data_final)
```

```python
new_col_headers = dem_data_final.columns.drop(['households']).insert(0,'< 5_years')
dem_data_final.columns=new_col_headers
display(dem_data_final)
```

### Concatenate and export

```python
zipcode_data_final = pd.concat([re_data_final, dem_data_final],axis = 1)
display(zipcode_data_final)
```
